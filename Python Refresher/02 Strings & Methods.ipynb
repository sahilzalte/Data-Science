# Python Strings and Methods

A short walkthrough of string basics, common methods, and slicing.

```python
name = "sahil"
```

## String Basics

Start with a simple string variable and inspect its value.

```python
name
```

```python
name.lower()
```

## String Methods

These methods return transformed values or useful information about the string.

```python
name.count("a")
```

```python
len(name)
```

## String Length and Slicing

Use indexing and slicing to pull out parts of a string.

```python
name[0:2]
```

```python
name[1:-1]
```

```python
name[1:12]
```
