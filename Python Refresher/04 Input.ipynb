# Python Input Methods

Learn how to take user input and process it in different formats.

## String Input

The `input()` function always returns a string. Use it when you want to capture text from the user.

```python
name = input("Enter your name: ")
```

```python
name
```

## Integer Input

Convert input to an integer using `int()` to perform numeric operations.

```python
a = int(input("Enter Value For a:"))
```

```python
b = int(input("Enter Value For b:"))
```

```python
a+b
```

## Summary

- Use `input()` to get user input as a string
- Use `int()`, `float()`, etc. to convert input to the desired type
- Always validate input before using it in calculations
