# String Formatting in Python

A short guide to common string formatting techniques: the `format()` method, f-strings, and alignment options.

```python
# Old Method
name = "Alice"
age = 25
print("My name is {} and I am {} years old.".format(name, age))
```

## Old-style: `str.format()`

The `str.format()` method works in all Python 3 versions and is useful when you need format placeholders.

```python
# New Method
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old.")
# Output: My name is Alice and I am 25 years old.
```

## Modern: f-strings

f-strings (Python 3.6+) are concise and support inline expressions.

```python
print(f"{'Python':<10}")  # Left-align
print(f"{'Python':>10}")  # Right-align
print(f"{'Python':^10}")  # Center-align
```

## Alignment and Width

Use format specifications inside f-strings to control alignment and width.

## Takeaway

- Prefer f-strings for readability and performance.
- Use `str.format()` when you need compatibility with older Python versions.
- Use format specifiers like `:<width>`, `:>width`, and `:^width` for alignment.
