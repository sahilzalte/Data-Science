## Basic If-Else

Use an `if` statement when you want to run code only if a condition is true.

```python
x = 10
if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but not more than 10")
elif x == 5:
    print("x is exactly 5")
else:
    print("x is less than 5")
```

## If-Elif-Else

Use `elif` when you want to test multiple conditions in order.

```python
def greet():
    return "hello"
```

## Functions

Functions let you group reusable logic under a clear, descriptive name.

```python
greet()
```

```python
def add_numbers(a, b):
    return a + b
```

```python
add_numbers(50, 60)
```
