# Match-Case in Python

Use `match`/`case` when you want to choose between several fixed values in a clear, readable way.

## HTTP Status Mapping

This example maps common HTTP status codes to readable messages.

```python
def get_http_status_message(code):
    match code:
        case 200:
            return "OK"
        case 400:
            return "Bad Request"
        case 404:
            return "Not Found"
        case 500:
            return "Internal Server Error"
        case _:
            return "Unknown Status"
```

```python
get_http_status_message(418)
```
